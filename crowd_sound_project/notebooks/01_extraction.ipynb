# Feature Extraction

This notebook extracts acoustic features from crowd sound stimuli using `librosa`.

**Features extracted per sound:**
- Duration
- Global intensity (RMS)
- F0 (pitch) mean and variability (YIN algorithm)
- Spectral centroid, bandwidth, rolloff
- MFCC (13 coefficients)

Results are saved as `outputs/<sound_name>/results.json`.

## Imports

```python
import os
import json
import numpy as np
import librosa
```

## Configuration

```python
audio_folder = "../data/sounds"
output_folder = "../outputs"
os.makedirs(output_folder, exist_ok=True)

print("Audio files found:", os.listdir(audio_folder))
```

## Feature Extraction Loop

```python
results = {}

for file in sorted(os.listdir(audio_folder)):
    if not file.endswith(".wav"):
        continue

    path = os.path.join(audio_folder, file)
    audio, sr = librosa.load(path, sr=None)
    print(f"\nProcessing: {file} ({len(audio)/sr:.2f}s)")

    # Duration
    duration = len(audio) / sr

    # Intensity (RMS)
    rms = librosa.feature.rms(y=audio)
    global_intensity = float(np.mean(rms))

    # F0 (pitch) via YIN
    f0 = librosa.yin(audio, fmin=50, fmax=500)
    f0_mean = float(np.nanmean(f0))
    f0_var = float(np.nanstd(f0))

    # Spectral features
    spectral_centroid = librosa.feature.spectral_centroid(y=audio, sr=sr)
    spectral_bandwidth = librosa.feature.spectral_bandwidth(y=audio, sr=sr)
    spectral_rolloff = librosa.feature.spectral_rolloff(y=audio, sr=sr)
    mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)

    # Build result dict
    result = {
        "file": file,
        "duration": float(duration),
        "global_intensity": global_intensity,
        "f0_mean": f0_mean,
        "f0_variability": f0_var,
        "spectral_centroid_mean": float(np.mean(spectral_centroid)),
        "spectral_bandwidth_mean": float(np.mean(spectral_bandwidth)),
        "spectral_rolloff_mean": float(np.mean(spectral_rolloff)),
        "mfcc_mean": mfcc.mean(axis=1).tolist()
    }

    # Save JSON
    sound_name = file.replace(".wav", "")
    sound_output_folder = os.path.join(output_folder, sound_name)
    os.makedirs(sound_output_folder, exist_ok=True)
    json_path = os.path.join(sound_output_folder, "results.json")
    with open(json_path, "w") as f:
        json.dump(result, f, indent=4)

    results[file] = result
    print(f"  Saved: {json_path}")

print("\nDONE — all files processed")
```

## Summary

```python
print(f"Total sounds processed: {len(results)}")
for name, r in results.items():
    print(f"  {name}: duration={r['duration']:.2f}s, F0={r['f0_mean']:.1f}Hz, intensity={r['global_intensity']:.4f}")
```
